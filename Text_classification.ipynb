---
การทำ feature extraction

---

```python
data = df['reviewText'].values.astype('U')
y = df.overall.values
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data)

print(X.shape)
print(X.toarray())
print(vectorizer.get_feature_names_out())
```

---
ทำ Logistic Regression
และแสดงค่าความแม่ยำ

---

```python
from sklearn.linear_model import LogisticRegression

X = X.toarray()

clf = LogisticRegression(random_state=0).fit(X[:5000], y[:5000])
```

```python
print(clf.predict(X[5000:5005]),"\n")
print(clf.predict_proba(X[5000:5005]),"\n")
#ค่าความแม่นยำของโมเดล
print(clf.score(X[5000:5005], y[5000:5005]))
```

---
ทำ Decision Tree
และแสดงค่าความแม่นยำ

---


```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn import metrics

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.35)

model = tree.DecisionTreeClassifier(max_depth=8)#กำหนดค่าความลึก = 8
model.fit(x_test, y_test)
y_pred = model.predict(x_test)

tree.plot_tree(model)
```

```python
#ค่าความแม่นยำของโมเดล 
print(metrics.f1_score(y_test,y_pred,average='weighted'))
```

---
ทำ Random Forest
และแสดงค่าความแม่นยำ

---

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.4) 
  
regressor = RandomForestRegressor(n_estimators = 10)
  
regressor.fit(x_train, y_train)  

y_pred = regressor.predict(x_test)
```

```python
#ค่าความแม่นยำของโมเดล 
print(regressor.score(x_test, y_pred))
```
